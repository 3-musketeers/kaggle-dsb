```python
import numpy as np

WORKING_PATH = "../../../../output/build-simple-model/"
final_images = np.load(WORKING_PATH + "trainImages.npy")
final_masks = np.load(WORKING_PATH + "trainMasks.npy")
```

```python
print(final_images[0,0])
print(final_images[1,0])
print(final_masks)
```

```python
import pylab
pylab.imshow(final_images[0,0], cmap=pylab.cm.bone) 
pylab.show()
pylab.imshow(final_masks[0,0], cmap=pylab.cm.bone) 
pylab.show()
```

```python
imgs = final_images[1,0]
lungmask = final_masks[1,0]

applied_mask = np.ndarray(final_images[1,0].shape)

for i in range(len(imgs)):
    applied_mask[i] = imgs[i] * lungmask[i]
pylab.imshow(applied_mask, cmap='gray') 
pylab.show()
```
