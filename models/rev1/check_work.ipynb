```python
import numpy as np
OUTPUT_PATH = '../../../output/rev1/'
lung_img = np.load(OUTPUT_PATH + 'lung_img_0000_0001.npz.npy')
```

```python
print(lung_img.shape)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

def plot_ct_scan(scan):
    f, plots = plt.subplots(int(scan.shape[0] / 20) + 1, 4, figsize=(25, 25))
    for i in range(0, scan.shape[0], 5):
        plots[int(i / 20), int((i % 20) / 5)].axis('off')
        plots[int(i / 20), int((i % 20) / 5)].imshow(scan[i], cmap=plt.cm.bone) 
```

```python
plot_ct_scan(lung_img)
```

```python
plt.imshow(lung_img[200], cmap=plt.cm.bone)
```

```python
lung_mask = np.load(OUTPUT_PATH + 'lung_mask_0000_0001.npz.npy')
```
